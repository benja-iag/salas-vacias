```python
def capa1(rooms):
	output = []
	for room in rooms:
		temp = {
			"edificio" : "V432",
			"piso" : -1,
			"sala" : room,
		}
		output.append(temp)
	return output
	
pprint(capa1(['SIM','LOC']))
```

```python
def capa2(rooms):
	output = []
	for i in range(0,len(rooms)-1,2):
		temp = {
			"edificio": "E441",
			"piso" : 5,
			"sala" : rooms[i] + " " + rooms[i+1],
		}
		output.append(temp)	
	return output


pprint(capa2(['LAB', 'INFORMATICA','LAB', 'TELEMATICA']))
```

```python
def capa3(rooms):
	output = []
	for i in range(0,len(rooms)-2,3):
		temp = {
			"edificio" : rooms[i],
			"piso" : rooms[i+1],
			"sala" : rooms[i+2],
		}
		output.append(temp)
	return output


pprint(capa3([
		'E441',
      '2',
      'S202',
      'E441',
      '2',
      'S206',
      'E441',
      '3',
      'S303',
]))
```

```python
#E441 4 L D
def capa4(rooms):
   output = []
   for i in range(0,len(rooms)-3,4):
      temp = {
         "edificio" : rooms[i],
         "piso" : rooms[i+1],
         "sala" : "LAB " + rooms[i+3],
      }
      output.append(temp)
   return output
```

```python
salas = []
salas = capa1(test['1'])
salas.extend(capa2(test['2']))
salas.extend(capa3(test['3']))
salas.extend(capa4(test['4']))
pprint(salas)
```

```python
# export to json
with open('salas.json', 'w') as f:
   json.dump(salas, f, indent=4)
```

```python
import copy
salas_edificios = {} #salas separadas por edificio

for sala in salas:
   edificio = sala['edificio']
   if edificio not in salas_edificios:
      salas_edificios[edificio] = {}
   piso = f'{sala['piso']}'
   if sala['piso'] not in salas_edificios[edificio]:
      salas_edificios[edificio][piso] = []
   salas_edificios[edificio][piso].append(sala['sala'])
   piso = f'{sala['piso']}'

# pprint(salas_edificios)



pprint(salas_edificios) 
```
